```python
import json
from googletrans import Translator

# Specify the path to your input and output JSON files
input_json_file_path = 'merged_tweets.json'
output_json_file_path = 'merged_tweets_translated.json'

# Read JSON data from the input file
with open(input_json_file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Initialize the translator
translator = Translator()

# Function to translate content
def translate_content(text, target_language='en'):
    translation = translator.translate(text, dest=target_language)
    return translation.text

# Translate the content in the JSON data
for username, user_data in data.items():
    for tweet in user_data.get("tweets", []):
        # Translate the content and update the JSON data
        translated_content = translate_content(tweet.get("content", ""))
        tweet["translated_content"] = translated_content

# Save the translated data to a new JSON file
with open(output_json_file_path, 'w', encoding='utf-8') as output_file:
    json.dump(data, output_file, ensure_ascii=False, indent=4)

print(f'Translation completed. Translated data saved to {output_json_file_path}')
```
