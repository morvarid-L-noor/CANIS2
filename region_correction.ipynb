```python
germany_counts = int(country_counts.get("Germany", 0) / country_counts.sum() * category_counts.get("EU", 0) + 1)
france_counts = int(country_counts.get("France", 0) / country_counts.sum() * category_counts.get("EU", 0) +1)
Netherlands_counts = int(country_counts.get("Netherlands", 0) / country_counts.sum() * category_counts.get("EU", 0)+1 )
Austria_counts = int(country_counts.get("Austria", 0) / country_counts.sum() * category_counts.get("EU", 0)+1 )
Spain_counts = int(country_counts.get("Spain", 0) / country_counts.sum() * category_counts.get("EU", 0)+1 )
Bulgaria_counts = int(country_counts.get("Bulgaria", 0) / country_counts.sum() * category_counts.get("EU", 0)+1 )
Finland_counts = int(country_counts.get("Finland", 0) / country_counts.sum() * category_counts.get("EU", 0) +1)

print(germany_counts, france_counts, Netherlands_counts, australia_counts, Spain_counts, Bulgaria_counts, Finland_counts)  # Adjust for all EU countries

eu_rows = df[df['Region of Focus'] == 'EU']

# Update 'EU' occurrences to the respective countries
eu_rows.loc[eu_rows.index[:germany_counts], 'Region of Focus'] = 'Germany'

eu_rows.loc[eu_rows.index[germany_counts:germany_counts+france_counts], 'Region of Focus'] = 'France'
print(df['Region of Focus'].value_counts())
print('------------------------------')

df.update(eu_rows)
eu_rows.loc[eu_rows.index[germany_counts+france_counts:germany_counts+france_counts+Netherlands_counts], 'Region of Focus'] = 'Netherlands'
print(df['Region of Focus'].value_counts())
print('------------------------------')

df.update(eu_rows)

eu_rows.loc[eu_rows.index[ germany_counts+france_counts+Netherlands_counts:germany_counts+france_counts+Netherlands_counts+australia_counts], 'Region of Focus'] = 'Austria'
print(df['Region of Focus'].value_counts())
print('------------------------------')

df.update(eu_rows)

eu_rows.loc[eu_rows.index[germany_counts+france_counts+Netherlands_counts+australia_counts:germany_counts+france_counts+Netherlands_counts+australia_counts+Spain_counts], 'Region of Focus'] = 'Spain'
print(df['Region of Focus'].value_counts())
print('------------------------------')

df.update(eu_rows)
eu_rows.loc[eu_rows.index[germany_counts+france_counts+Netherlands_counts+australia_counts+Spain_counts: germany_counts+france_counts+Netherlands_counts+australia_counts+Spain_counts+Bulgaria_counts], 'Region of Focus'] = 'Bulgaria'

print(df['Region of Focus'].value_counts())
print('------------------------------')
index = germany_counts+france_counts+Netherlands_counts+australia_counts+Spain_counts + Bulgaria_counts

df.update(eu_rows)
eu_rows.loc[eu_rows.index[index: index+france_counts], 'Region of Focus'] = 'Belgium'
eu_rows.loc[eu_rows.index[index+france_counts: index+2*france_counts], 'Region of Focus'] = 'Greece'
eu_rows.loc[eu_rows.index[index+2*france_counts: index+3*france_counts], 'Region of Focus'] = 'Poland'
eu_rows.loc[eu_rows.index[index+3*france_counts: index+4*france_counts], 'Region of Focus'] = 'Italy'
index = index+4*france_counts
eu_rows.loc[eu_rows.index[index: index+Bulgaria_counts], 'Region of Focus'] = 'Malta'
eu_rows.loc[eu_rows.index[index+Bulgaria_counts: index+2*Bulgaria_counts], 'Region of Focus'] = 'Portugal'
eu_rows.loc[eu_rows.index[index+2*Bulgaria_counts: index+3*Bulgaria_counts], 'Region of Focus'] = 'Lithuania'
index = index+3*Bulgaria_counts
eu_rows.loc[eu_rows.index[index: index+Finland_counts], 'Region of Focus'] = 'Finland'
eu_rows.loc[eu_rows.index[index+Finland_counts: index+2*Finland_counts], 'Region of Focus'] = 'Cyprus'
eu_rows.loc[eu_rows.index[index+2*Finland_counts: index+3*Finland_counts], 'Region of Focus'] = 'Croatia'
eu_rows.loc[eu_rows.index[index+3*Finland_counts: index+4*Finland_counts], 'Region of Focus'] = 'Slovakia'
eu_rows.loc[eu_rows.index[index+4*Finland_counts: index+5*Finland_counts], 'Region of Focus'] = 'Denmark'
eu_rows.loc[eu_rows.index[index+5*Finland_counts: index+6*Finland_counts], 'Region of Focus'] = 'Hungary'
eu_rows.loc[eu_rows.index[index+6*Finland_counts: index+7*Finland_counts], 'Region of Focus'] = 'Estonia'
eu_rows.loc[eu_rows.index[index+7*Finland_counts: index+8*Finland_counts], 'Region of Focus'] = 'Latvia'
eu_rows.loc[eu_rows.index[index+8*Finland_counts: index+9*Finland_counts], 'Region of Focus'] = 'Romania'

print(df['Region of Focus'].value_counts())
print('------------------------------')

df.update(eu_rows)

# Check the modified DataFrame
print(df['Region of Focus'].value_counts())

```
